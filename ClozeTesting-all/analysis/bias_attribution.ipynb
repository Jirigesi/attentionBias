Goal: to find the statistical significance of the bias in the data
Data: mispredicted data, correct predicted data

What we can comapre:
1. The average arribution scores betwenn the two groups
2. Top one attribution scores between the two groups
3. The average of top 5 attribution scores between the two groups
4. The average of top 10 attribution scores between the two groups

There could be two kinds of data seperations:

   1. **Attribution score based**: Collect all attribution scores on each syntax in one large list. And after comparing the top attribution score between correct and mispredicted data, I found there is a statitical significance. But, the problem is that, the attribution scores are not saved for each instance. For example attribution on KeyWord, some instances could have multiple KeyWords. 
   2. **Instance based**: Save the attribution score on each syntax for each instance. And then, comparing the top attribution scores based on instances within correct and mispredicted groups.

```python
def Avg_attribution_ttest(correct_data, mispredict_data):
    # get the average of layer and head for SEP token
    correct_data_avg = np.mean(correct_data, axis=2)
    mispredict_data_avg = np.mean(mispredict_data, axis=2)
    # calculate the sum of each layer and head for SEP token
    correct_data_avg_sum = np.sum(correct_data_avg, axis=1)
    mispredict_data_avg_sum = np.sum(mispredict_data_avg, axis=1)
    # calculate the t-test for each layer and head
    result = ttest_ind(correct_data_avg_sum, mispredict_data_avg_sum)
    print(result)
    
```

```python
def Top1_attribution_ttest(correct_data, mispredict_data):
    # calculate the top one attribution score for each layer and head
    top1_correct = np.max(correct_data, axis=2)
    top1_mispredict = np.max(mispredict_data, axis=2)
    # calculate the sum of each layer and head for SEP token
    top1_correct_sum = np.sum(top1_correct, axis=1)
    top1_mispredict_sum = np.sum(top1_mispredict, axis=1)
    # calculate the t-test for each layer and head
    result = ttest_ind(top1_correct_sum, top1_mispredict_sum)
    print(result)
```

```python
def TopAvg_attribution_test(correct_data, mispredict_data, top=5):
    # calculate the top five attribution score for each layer and head
    top5_correct = np.sort(correct_data, axis=2)[:, :, -top:]
    top5_mispredict = np.sort(mispredict_data, axis=2)[:, :, -top:]
    # calculate the avgerge of selected top five attribution score for each layer and head
    top5avg_correct = np.mean(top5_correct, axis=2)
    top5avg_mispredict = np.mean(top5_mispredict, axis=2)
    # calculate the sum of each layer and head for SEP token
    top5avg_correct_sum = np.sum(top5avg_correct, axis=1)
    top5avg_mispredict_sum = np.sum(top5avg_mispredict, axis=1)
    # calculate the t-test for each layer and head
    result = ttest_ind(top5avg_correct_sum, top5avg_mispredict_sum)
    print(result)
```

```python
def Extract_df4boxplot(syntax_type, top_attrition):
        print("Syntax type: ", syntax_type)
        
        syntax_type_correct = np.array(syntax_correct[syntax_type])
        syntax_type_mispredict = np.array(syntax_mispredict[syntax_type])

        df = pd.DataFrame(columns=['layer', 'prediction', 'attribution'])

        for layer_number in range(12):
                # change the shape of the data from (layer, head, token) to (layer, token)
                layer_type_correct = np.reshape(syntax_type_correct, (12, 12*syntax_type_correct.shape[2]))
                # get top attribution score for specific layer 
                layer_top_correct = np.sort(layer_type_correct[layer_number, :])[-top_attrition:]
                dict = {'layer': [layer_number] * len(layer_top_correct), 
                        'prediction': ['correct'] * len(layer_top_correct), 
                        'attribution': layer_top_correct}
                tmp_df = pd.DataFrame(dict)
                df = pd.concat([df, tmp_df], ignore_index=True)

                # change the shape of the data from (layer, head, token) to (layer, token)
                layer_type_mispredict = np.reshape(syntax_type_mispredict, (12, 12*syntax_type_mispredict.shape[2]))
                # get top 2000 attribution score for for layer one for mispredict
                layer_top_mispredict = np.sort(layer_type_mispredict[layer_number, :])[-top_attrition:]

                dict = {'layer': [layer_number] * len(layer_top_mispredict), 
                        'prediction': ['mispredict'] * len(layer_top_mispredict), 
                        'attribution': layer_top_mispredict}
                tmp_df = pd.DataFrame(dict)
                df = pd.concat([df, tmp_df], ignore_index=True)

        return df
```

#### Layer-level attribution score averge t-test

```python
def get_instance_top1_syntax_layer_avg_attribution(syntax_type, instance_correct, instance_mispredict):
    annotation_correct = instance_correct[syntax_type]
    annotation_mispredict = instance_mispredict[syntax_type]
    top_correct_annotation = []

    for instance in annotation_correct:
        instance = np.array(instance)
        top_correct_annotation.append(np.max(instance, axis=2))
        
    top_mispredict_annotation = []

    for instance in annotation_mispredict:
        instance = np.array(instance)
        top_mispredict_annotation.append(np.max(instance, axis=2))
        
    top_mispredict_annotation = np.array(top_mispredict_annotation)
    top_correct_annotation = np.array(top_correct_annotation)

    # convert top_mispredict_annotation from length, layers, heads to layers, heads, length

    coverted_top_mispredict_annotation = np.transpose(top_mispredict_annotation, (1, 2, 0))

    coverted_top_correct_annotation = np.transpose(top_correct_annotation, (1, 2, 0))
    
    avg_correct_annotation = np.mean(coverted_top_correct_annotation, axis=2)
    avg_mispredict_annotation = np.mean(coverted_top_mispredict_annotation, axis=2)
    
    return avg_correct_annotation, avg_mispredict_annotation
```

```python
for syntax_type in instance_correct.keys():
    if syntax_type != 'booleandecimalinteger':
        print("Syntax type: ", syntax_type)
        avg_correct_annotation, avg_mispredict_annotation = get_instance_top1_syntax_layer_avg_attribution(syntax_type, instance_correct, instance_mispredict)
        # ttest for each layer 
        significantlayer = [] 
        for i in range(12):
            p_value = ttest_ind(avg_correct_annotation[i], avg_mispredict_annotation[i]).pvalue
            if p_value < 0.05:
                significantlayer.append(i)
                
        if len(significantlayer) == 0:
            print("No significant layer")
        else:
            print("Significant layer: ", significantlayer)
```

From the t-test result, we can see that, there is no statistical significance between the two groups for all syntax types on all layer level mean attribution scores.

```python
import numpy as np
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
def get_syntax_head_level_attribution_ttest_95p_result(syntax_type, instance_correct, instance_mispredict):
    annotation_correct = instance_correct[syntax_type]
    annotation_mispredict = instance_mispredict[syntax_type]
    top_correct_annotation = []

    for instance in annotation_correct:
        instance = np.array(instance)
        top_correct_annotation.append(np.max(instance, axis=2))
        
    top_mispredict_annotation = []

    for instance in annotation_mispredict:
        instance = np.array(instance)
        top_mispredict_annotation.append(np.max(instance, axis=2))
        
    top_mispredict_annotation = np.array(top_mispredict_annotation)
    top_correct_annotation = np.array(top_correct_annotation)

    # convert top_mispredict_annotation from length, layers, heads to layers, heads, length

    coverted_top_mispredict_annotation = np.transpose(top_mispredict_annotation, (1, 2, 0))
    coverted_top_correct_annotation = np.transpose(top_correct_annotation, (1, 2, 0))

    p_value_distribution = []
    significant_head_number = 0

    for layer in range(coverted_top_mispredict_annotation.shape[0]):
        layer_attronutions = []
        for head in range(coverted_top_mispredict_annotation.shape[1]):
            # get 95% confidence interval
            correct_mean, correct_lower, correct_upper = mean_confidence_interval(coverted_top_correct_annotation[layer][head])
            mispred_mean, mispred_lower, mispred_upper = mean_confidence_interval(coverted_top_mispredict_annotation[layer][head])
            # get data between the lower and upper bound
            P95_correct_data = [x for x in coverted_top_correct_annotation[layer][head] if x >= correct_lower and x <= correct_upper]
            P95_mispred_data = [x for x in coverted_top_mispredict_annotation[layer][head] if x >= mispred_lower and x <= mispred_upper]
            
            p_value = ttest_ind(P95_correct_data, P95_mispred_data).pvalue
            if p_value < 0.05:
                layer_attronutions.append(0.05)
                significant_head_number += 1
            else:
                layer_attronutions.append(1)
        p_value_distribution.append(layer_attronutions)
        
    return p_value_distribution, significant_head_number
```

```python
def get_syntax_head_level_attribution_95_top_heads(syntax_list, instance_correct, instance_mispredict):
    significant_head_record = [[[] for col in range(12)] for row in range(12)]
    
    for syntax_type in syntax_list:
        annotation_correct = instance_correct[syntax_type]
        annotation_mispredict = instance_mispredict[syntax_type]
        
        top_correct_annotation = []
        for instance in annotation_correct:
            instance = np.array(instance)
            top_correct_annotation.append(np.max(instance, axis=2))
            
        top_mispredict_annotation = []
        for instance in annotation_mispredict:
            instance = np.array(instance)
            top_mispredict_annotation.append(np.max(instance, axis=2))
            
        top_mispredict_annotation = np.array(top_mispredict_annotation)
        top_correct_annotation = np.array(top_correct_annotation)

        # convert top_mispredict_annotation from length, layers, heads to layers, heads, length

        coverted_top_mispredict_annotation = np.transpose(top_mispredict_annotation, (1, 2, 0))
        coverted_top_correct_annotation = np.transpose(top_correct_annotation, (1, 2, 0))

        for layer in range(coverted_top_mispredict_annotation.shape[0]):
            layer_attronutions = []
            for head in range(coverted_top_mispredict_annotation.shape[1]):
                # get 95% confidence interval
                correct_mean, correct_lower, correct_upper = mean_confidence_interval(coverted_top_correct_annotation[layer][head])
                mispred_mean, mispred_lower, mispred_upper = mean_confidence_interval(coverted_top_mispredict_annotation[layer][head])
                # get data between the lower and upper bound
                P95_correct_data = [x for x in coverted_top_correct_annotation[layer][head] if x >= correct_lower and x <= correct_upper]
                P95_mispred_data = [x for x in coverted_top_mispredict_annotation[layer][head] if x >= mispred_lower and x <= mispred_upper]
                p_value = ttest_ind(P95_correct_data, P95_mispred_data).pvalue
                # p_value = ttest_ind(coverted_top_mispredict_annotation[layer][head], coverted_top_correct_annotation[layer][head]).pvalue
                if p_value < 0.05:
                    significant_head_record[layer][head].append(syntax_type)
        
    return significant_head_record
```
